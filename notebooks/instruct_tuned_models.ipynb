# Instruction-Tuned Models

## What are Instruction-Tuned Models: --

Instruction-tuned models are language models that have been **trained to follow instructions** instead of just completing text.

---

## Base Model vs Instruction-Tuned Model

### Base Model (e.g., Llama 3.1 Base)
- **Trained on**: Books, websites, code (just predicting next word)
- **Good at**: Completing sentences, generating text that flows naturally
- **Problem**: Doesn't understand you're asking a question

**Example:**
```
Input: "What is the capital of France?"
Output: "is a common question asked in geography classes..."
```
❌ It just continues the text, doesn't answer!

### Instruction-Tuned Model (e.g., Llama 3.1 Instruct)
- **Extra training on**: Question-answer pairs, instructions with responses
- **Good at**: Following instructions, answering questions, staying on topic
- **Understands**: When you ask a question, you want an answer

**Example:**
```
Input: "What is the capital of France?"
Output: "The capital of France is Paris."
```
✅ It understands and answers directly!

---

## How Instruction Tuning Works

```
Step 1: Pre-training (Base Model)
├── Read millions of web pages, books, code
├── Learn language patterns and facts
└── Result: Can generate text, but doesn't follow instructions

Step 2: Instruction Fine-Tuning
├── Train on thousands of (instruction → response) pairs
│   Example pairs:
│   • "Explain Python" → "Python is a programming language..."
│   • "Summarize this code" → [concise summary]
│   • "Answer based on context" → [uses only provided info]
└── Result: Follows instructions like a helpful assistant
```
